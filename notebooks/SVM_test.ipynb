```python
import numpy as np
import statsmodels.api as sm
import cPickle as pickle
import GPy
import GPy, numpy as np
import theano.tensor as T, theano
from theano.tensor import slinalg, nlinalg

from matplotlib import pyplot as plt
%matplotlib inline
plt.ion()
```

```python
n, p, d = 100, 2, 1

X = np.random.uniform(-1, 1, (n, p))
Y = np.cos(X[:, [0]]) + np.sin(X[:, [1]])**2 + np.random.normal(0, .1, (n, d))
Y -= Y.mean()
Y = np.where(Y>0, -1, 1)
```

```python
plt.scatter(*X[Y[:, 0]==1].T, facecolors='none', edgecolors='k', linewidth=.3)
plt.scatter(*X[Y[:, 0]==-1].T, facecolors='k', edgecolors='k')
```

```python
class SVM(GPy.Model):
    def __init__(self, X, Y, kernel=None, C=.01, name='SVM'):
        super(SVM, self).__init__(name=name)
        if kernel is None:
            from SymDiffPy.kernel.rbf import RBF
            kernel = RBF(X.shape[1])
        self.kernel = kernel
        
        self.X = X
        self.Y = Y
        
        self.alpha = GPy.Param('alpha', np.ones(self.X.shape[0]), GPy.constraints.Logistic(0, C))
        self.link_parameter(self.alpha)
        self.link_parameter(self.kernel)
        
        self.T_X1 = T.dmatrix('X1')
        self.T_X2 = T.dmatrix('X2')
        self.T_alpha = T.dvector('alpha')

        self.kernel.init_theano(self.T_X1, self.T_X2)
        
        self.T_obj = T.sum(self.T_alpha) - .5*T.sum(T.outer(self.T_alpha, self.T_alpha) * self.Y.dot(self.Y.T) * self.kernel.T_K)
        self.T_grad = theano.grad(self.T_obj, [self.T_alpha] + self.kernel.T_list)
        
        self.f_obj = theano.function([self.T_alpha, self.T_X1, self.T_X2] + self.kernel.T_list, [self.T_obj, self.kernel.T_K])
        self.f_grad = theano.function([self.T_alpha, self.T_X1, self.T_X2] + self.kernel.T_list, self.T_grad)
        
    def parameters_changed(self):
        param_list = [self.alpha, self.X, self.X] + self.kernel.flattened_parameters
        self._obj, self.K = self.f_obj(*param_list)
        self.gradient[:] = np.concatenate(self.f_grad(*param_list))
        
    def log_likelihood(self):
        return self._obj
    
    def predict(self, Xnew):
        mu = self.alpha.dot(self.K).dot(self.Y)
        return mu
    
    def plot(self):
        plt.scatter(*X[Y[:, 0]==1].T, facecolors='none', edgecolors='k', linewidth=.3)
        plt.scatter(*X[Y[:, 0]==-1].T, facecolors='k', edgecolors='k')
```

```python
svm = SVM(X, Y)
```

```python
svm.optimize(messages=1, ipython_notebook=1)
```
