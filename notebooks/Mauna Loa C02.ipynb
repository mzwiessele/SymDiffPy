```python
with open('mauna_data.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
X = np.array(data.time)[:, None]
Y = np.array(data.co2)[:, None]
```

```python
from SymDiffPy.kernel.rbf import RBF
from SymDiffPy.kernel.periodic import PeriodicExp
from SymDiffPy.kernel.linear import Linear
from SymDiffPy.kernel.sum import Sum
from SymDiffPy.models.gp import GP
import numpy as np, GPy

%matplotlib inline
from matplotlib import pyplot as plt
```

```python
m = GP(X, Y, Sum([RBF(1, 66, lengthscale=67, name='long'), PeriodicExp(1), Linear(1)]))
```

```python
m.constrain_positive()
m.kernel.periodicexponential.period.fix()
```

```python
m.noise = .2
```

```python
m.optimize('bfgs',messages=1, ipython_notebook=1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18,6))
ax.scatter(X, Y, color='k', marker='x')
mi, ma = X.min(), X.max()
ten_perc = .1*(ma-mi)
Xpred = np.linspace(mi, 2021, 500)[:, None]
mu, var = m.predict(Xpred)
#ax.plot(Xpred, mu, color='g', lw=1.5)
ax.fill_between(Xpred[:, 0], mu[:,0]+2*np.sqrt(np.diagonal(var)), mu[:,0]-2*np.sqrt(np.diagonal(var)), color='k', alpha=.1)
ax.set_xlabel('year')
ax.set_ylabel('CO2 concentration [ppm]')
```
